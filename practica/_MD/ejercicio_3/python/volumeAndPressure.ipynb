```python
import pandas as pd
from matplotlib import pyplot as plt
from lib import get_filepath, get_thermodinamic_df, get_pressure
```

```python
temperature = 300
volumes = tuple(range(10**6, 9*10**6, 10**6))
pressures = []
sufix = '__termodinamica'

for volume in volumes:
  filepath = get_filepath(temperature, volume, sufix)
  thermodinamic_df = get_thermodinamic_df(filepath)
  pressure = get_pressure(thermodinamic_df)
  pressures.append(pressure)
```

```python
df = pd.DataFrame({'Volumen [ang**3]': volumes, 'Presion [Atm]': pressures})
df.plot(x='Volumen [ang**3]', y='Presion [Atm]')

plt.savefig('../dat/volumeAndPressure.png')
```
